```python
import os
import cv2
import json 


def get_files(path,add_path = True, pattern_ = None, remove = None, sort_ = True):
    if os.path.exists(path) == False:
        return []

    files_ =  os.listdir(path)

    if ".DS_Store" in files_:
        files_.remove(".DS_Store")

    if "._.DS_Store" in files_:
        files_.remove("._.DS_Store")

    if remove is not None:
        files_ = [file_ for file_ in files_ if remove not in file_]

    if pattern_ != None:
        files_ = [file_ + pattern_ for file_ in files_]

    if add_path:
        return [path + file_ for file_ in files_]
    else:
        return files_

def get_dataset_(path_,pattern_ = "/images/", limit_ = None, limit_view_ = None):
    files_ = get_files(path_)
    references_ = []
    
    for l_ , f_ in enumerate(files_):
        
        # open json file
        f = open (f_ + "/data.json", "r")
        data_json_ = json.loads(f.read())
        data_ = {'file' : f_ + "/image.jpeg" , 'categorie' : data_json_['categorie'] ,'label' :l_, 'query' : [] }
        
        t = get_files(f_ + pattern_)
        for id__ , view_ in enumerate(t):
            if "_mask" in view_:
                continue
            data_['query'].append({'file' : view_ , 'label' : l_ })
            if limit_view_ != None:
                if id__ == limit_view_ :
                    break
        references_.append(data_)
        if limit_ != None:
            if l_ == limit_ :
                break
                
    return references_

def dataset_to_dict(data_):
    types_ = {}
    for r in data_:
        if r['categorie'] not in types_:
            types_[r['categorie']] = []
        types_[r['categorie']].append(r)
    return types_



dataset_ = get_dataset_("./CBDIR/")
dict_ = dataset_to_dict(dataset_)
```
